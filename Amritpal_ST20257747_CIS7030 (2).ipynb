```python
#import Library
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt


# csv file naming
df = "Dataset/API_AG.YLD.CREL.KG_DS2_en_csv_v2_5734359.csv"
df2 = "Dataset/API_SP.POP.TOTL_DS2_en_csv_v2_4485025.csv"
df3 = 'Dataset/Bitcoin_tweets.csv'

# Reading CSV files
cereal_yield_data = pd.read_csv(df, skiprows=4)
population_data = pd.read_csv(df2, skiprows=4)
bitcoin_social_media_posts = pd.read_csv(df3)


print(cereal_yield_data)

cereal_output_recent_years = cereal_yield_data[['Country Code', '2019', '2020']]

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

global_cereal_production = world.merge(cereal_output_recent_years, left_on='iso_a3', right_on='Country Code')

global_cereal_production
```

```python
##1.2.4 Plot (scatter or line plot) the percentage change in cereal yield from 2011 to 2021, for the country
##      having the highest population in 2021.

top_population_country = world_population.sort_values('2021', ascending=False).iloc[0]

pop_title = top_population_country['Country Code']

top_country_cereal_output = cereal_yield_data[cereal_yield_data['Country Code'] == pop_title]

cy_2011 = top_country_cereal_output.iat[0, top_country_cereal_output.columns.get_loc('2011')]
cy_2021 = top_country_cereal_output.iat[0, top_country_cereal_output.columns.get_loc('2021')]

percentage_change = top_country_cereal_output[['2011', '2021']].pct_change(axis=1)['2021'].values[0] * 100

years = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
cereal_yield_values = top_country_cereal_output[years]

percentage_change_yearly = cereal_yield_values.pct_change(axis=1) * 100

plt.figure(figsize=(12, 6))
plt.plot(years, percentage_change_yearly.values.flatten(), marker='o', color='blue', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Percentage Change in Cereal Yield (%)')
plt.title(f'Percentage Change in Cereal Yield for {pop_title} (2011-2021)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
##1.2.5 Present a scatter plot between the mean population of each country and the mean cereal yield from
##    the year 2011 until 2021.
import numpy as np
import matplotlib.pyplot as plt

average_population = population_data.filter(items=['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']).mean(axis=1)
average_cereal_production = cereal_yield_data.filter(items=['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']).mean(axis=1)

data_list = []

for country, population, yield_ in zip(population_data['Country Name'], average_population, average_cereal_production):
    data_dict = {
        'Country Name': country,
        'Mean Population': population,
        'Mean Cereal Yield': yield_
    }
    data_list.append(data_dict)

plt.figure(figsize=(12, 8))
for data_point in data_list:
    plt.scatter(
        data_point['Mean Population'],
        data_point['Mean Cereal Yield'],
        alpha=0.7,
        c='orange',
        edgecolors='black',
        s=50,
        marker='o',
        label=data_point['Country Name']
    )

plt.title('Mean Pop vs Mean Cereal Yield (2011-2021)')
plt.xlabel('Mean Pop (2011-2021)')
plt.ylabel('Mean Cereal Yield (2011-2021)')
plt.xscale('log')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
import re

sampled_bitcoin_tweets = bitcoin_social_media_posts.sample(n=500)

sampled_bitcoin_tweets = sampled_bitcoin_tweets.dropna(subset=['text'])

sampled_bitcoin_tweets['clean_tweets'] = sampled_bitcoin_tweets['text'].apply(lambda text:
    re.sub(r'http\S+', '', text)
    .strip()
    .lower()
    .replace('[^\w\s]', '')
    .replace('\n', ' ')
    .replace('\r', ' ')
    .replace('\t', ' ')
    .replace('\s+', ' ')
)

sampled_bitcoin_tweets
```

```python
##2.2
from geopy.geocoders import Nominatim
from textblob import TextBlob
import pandas as pd
import random

geolocator = Nominatim(user_agent="geoencoding")

tweet_locations = []
tweet_sentiments = []

for index, row in sampled_bitcoin_tweets.iterrows():
    loc = row['user_location']
    cleaned_text = row['clean_tweets']

    # Geocode location
    if pd.notna(loc) and loc.strip():
        try:
            geo = geolocator.geocode(loc)
            if geo:
                tweet_locations.append((geo.latitude, geo.longitude))
            else:
                tweet_locations.append(None)
        except:
            tweet_locations.append(None)
    else:
        tweet_locations.append(None)


    analysis = TextBlob(cleaned_text)
    tweet_sentiments.append(analysis.sentiment)

sampled_bitcoin_tweets['coordinates'] = tweet_locations
sampled_bitcoin_tweets['sentiment'] = tweet_sentiments

sampled_bitcoin_tweets
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

geo_valid_tweets = sampled_bitcoin_tweets.dropna(subset=['coordinates', 'clean_tweets']).copy()

geo_valid_tweets['lat'] = [coord[0] for coord in geo_valid_tweets['coordinates']]
geo_valid_tweets['lon'] = [coord[1] for coord in geo_valid_tweets['coordinates']]

geo_valid_tweets['polarity'] = geo_valid_tweets['sentiment'].apply(lambda x: x.polarity)
geo_valid_tweets['subjectivity'] = geo_valid_tweets['sentiment'].apply(lambda x: x.subjectivity)
```

```python
##2.3
# Polarity Plot
fig, ax = plt.subplots(figsize=(10, 6))
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(ax=ax, color='lightgrey')

geo_valid_tweets.plot(kind='scatter', x='lon', y='lat',
                     c=geo_valid_tweets['polarity'], cmap='RdYlGn',
                     colorbar=True, alpha=0.5, ax=ax)

ax.set_title('Tweet Polarity Geospatial Distribution')
plt.show()
```

```python
#2.4
# Subjectivity Plot
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, color='lightgrey')

geo_valid_tweets.plot(kind='scatter', x='lon', y='lat',
                     c=geo_valid_tweets['subjectivity'], cmap='Blues',
                     colorbar=True, alpha=0.5, ax=ax)

ax.set_title('Tweet Subjectivity Geospatial Distribution')
plt.show()
```
